```python
df['image'] = image_names
df['image'] = df['image'].str.replace('assets_for_dr_training', 'images', regex=False)
df['class'] = 0
df
```

```python
df = df.sort_values(by='image', ascending=True).reset_index(drop=True)
df.head()
```

```python
df2 = df.copy()
df2.head()
```

```python
df2['digit'] = df2['image'].str.extractall(r'(\d+)').unstack().astype(int)
df2['digit'] = df2['digit'].apply(pd.to_numeric)
df2 = df2.sort_values(by='digit', ascending=True).reset_index(drop=True)
df2 = df2.drop('digit', axis=1)
df2.head(100)
```
